```python
import numpy as np 
```

```python
import pandas as pd
```

```python
short_pos = open("pos.txt","r").read()
short_neg = open("neg.txt","r").read()
```

```python
documents = []
documents1 = []
documents2 = []


for r in short_pos.split('\n'):
    documents1.append( r )
    documents.append( r )

for r in short_neg.split('\n'):
    documents2.append( r )
    documents.append( r )
```

```python
X = df.iloc[:, :18361].values
y = df.iloc[:,-1].values
```

```python
#splitting of data into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
y_test
```

```python
# #feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
#fit your training data to NB
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
#predict
y_predictor = classifier.predict(X_test)
```

```python
y_predictor
```

```python
#detect things using confusion matrix
from sklearn.metrics import confusion_matrix
falseCaci = confusion_matrix(y_test, y_predictor)
```

```python
falseCaci
```

```python
 from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predictor, normalize=True, sample_weight=None)
```
